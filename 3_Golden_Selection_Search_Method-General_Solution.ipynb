# Golden Selection Search Method - General Solution

```python
itr=int(input('Enter number of iterations: '))
print('')

aw=0
bw=1
i=0

#w=0.618
#w, a, b = 0.618, aw, bw
#x=(w*(b-a))+a

while i<itr:

	k=i+1

	lw=bw-aw
	u=aw+(0.618*lw)
	v=bw-(0.618*lw)

	zu=(25*(u**2))+(54/(5*u))
	zv=(25*(v**2))+(54/(5*v))

	print('Iteration ',k,':')
	print('W1 = ',u)
	print('W2 = ',v,'\n')
		
	if zu<zv:

		aw = v

	else:

		bw = u

	i+=1;
```
