# Exhaustive search bracketing

```python
a = 0
b = 5
dn = 0.5
i = 1
x = a + dn

while (i>=0):

    j=1

    x2 = x + dn
    x3 = x2 + dn
    x4 = x3 + dn
    
    y1 = x**2 + 54/x
    y2 = x2**2 + 54/x2
    y3 = x3**2 + 54/x3
    y4 = x4**2 + 54/x4

    print('Iteration {}:'.format(i))
    print('y1 =',y1)
    print('y2 =',y2)
    print('y3 =',y3)
    print('(x1, x3) = ({}, {})'.format(x,x3),'\n')

    if (y2 < y3):
        print('Solution has converged between (x1, x3) = ({}, {})'.format(x,x3),'\n')
        break;
    
    i += 1

    if (i==1000):
        print('Solution has reached {} iterations'.format(i))
        break;
    
    x = x2
```
