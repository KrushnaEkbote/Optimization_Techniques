```python
import math
l = 0.1
dl = 0.01
e = 0.01
i = 1

print('\ne =',e)
print ('')

while(i>0):
    f = 0.65 - (0.75/(1 + l**2)) - (0.65*l*math.atan(1/l))
    
    p = l-dl
    q = l+dl
    
    u = 0.65-(0.75/(1+p**2))-(0.65*p*math.atan(1/p))
    v = 0.65-(0.75/(1+q**2))-(0.65*q*math.atan(1/q))
    
    print('Iteration {}: '.format(i))
    print(' l{} ='.format(i),l)
    print(' f{} = {}'.format(i,f))
    #print(' f{}+ = {}\n f{}- = {}'.format(i,q,i,p))
    
    L = l-(dl*(v-u)/(2*(v-(2*f)+u)))
    
    dfl=abs((v-u)/(2*dl))

    print(' dfl =',dfl)
    
    if dfl<e:
        print('\nSoultion has converged at l{} ='.format(i),l,'\n')
        break
        
    if i>=500:
        print('\nSolution has reached',i,'iterations.\n')
        break

    l=L
    print('')
    i += 1
```
